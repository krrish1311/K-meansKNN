```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('weblog.csv')
```

```python
dataset
```

```python
dataset.columns
```

```python
check = dataset['Staus']

status=[]
```

```python

check.values.shape
```

```python
for i in check.values:
    if i == 'No' or 'dumped':
        status.append(404)
        
    else:
        status.append(int(i))
```

```python
status=np.array(status)
status=pd.DataFrame(status,columns=['Status'])   
```

```python
status
```

```python
from sklearn.preprocessing import LabelEncoder

label = LabelEncoder()

Ip = label.fit_transform(dataset['IP'])

Time = label.fit_transform(dataset['Time'])

URL = label.fit_transform(dataset['URL'])

Ds =pd.DataFrame(Ip,columns=['IP'])

Ds['Time']=pd.DataFrame(Time)

Ds['URL']=pd.DataFrame(URL)

Ds['Status']=status
```

```python
Ds
```

```python
status
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,50):
    mind=KMeans(n_clusters=i)
    mind.fit(Ds.values)
    w=mind.inertia_
    wcss.append(w)
```

```python
wcss[0:15]
```

```python
plt.scatter(range(1,50),wcss,marker='o')
```

```python
Ds
```

```python
model = KMeans(n_clusters=9)
```

```python
model.fit(Ds.values)


```

```python
pred=model.fit_predict(Ds.values)
```

```python
Ds['pred']=pred
```

```python
plt.scatter(Ds['Time'],Ds['Status'],c=Ds['pred'])
```
